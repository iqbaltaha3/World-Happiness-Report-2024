```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("whr.csv")
df.head()
```

The 2024 World Happiness Dataset offers insights into the relative happiness rankings of different population groups within each country compared to their counterparts worldwide. Each entry provides information on the happiness position of various demographic segments, shedding light on how they fare in comparison to similar groups in other countries.

For instance, taking Finland as an example, the dataset reveals that its young population ranks 7th globally in happiness compared to youth populations in other countries. Additionally, both the lower middle and upper middle classes of Finland top the global happiness rankings within their respective categories. However, Finland's elderly population comes in as the 1st runner-up, trailing behind Denmark's elderly population, which claims the top spot.

Two additional columns, happiest and least_happy, highlight the population groups within each country that experience the highest and lowest levels of happiness, respectively.

The all_ages column provides an overall ranking of each country based on the collective happiness levels of its population. This dataset serves as a valuable resource for understanding the happiness dynamics within and across nations, offering nuanced insights into the well-being of different demographic groups on a global scale.

```python
# Function to Extract a Specific country from the dataset

class CountryDataFrameExtractor:
    def __init__(self, dataframe):
        self.dataframe = dataframe
    
    def extract_country_data(self, country_name):
        
        filtered_df = self.dataframe[self.dataframe["country"] == country_name]
        return filtered_df
    
    
# Function to show a seaborn barplot

def set_plot_title(df):
    if len(df) != 1:
        print("Incorrect Name")
        return

    country_name = df.iloc[0]["country"]
    
    # Set the index of the row
    index = df.index[0]
    
    sns.barplot(df,palette = "pastel")
   
    plt.ylabel("World Happiness Ranking")
    plt.xlabel("Population Group")
    plt.title(f"Rank {index} :: {country_name}")
    plt.grid()
    plt.show()    
    

#Creating an object of CountryDataFrameExtractor class   
    
extractor = CountryDataFrameExtractor(df)    
```

This function combines above two functions and makes country wise analysis easier

```python
def Country(name):
    set_plot_title(extractor.extract_country_data(name))
```
